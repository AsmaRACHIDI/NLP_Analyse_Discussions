```python
import requests
import csv
import pandas as pd
```

```python
# Effectuer une requête GET pour récupérer les discussions de data.gouv.fr
response = requests.get("https://www.data.gouv.fr/api/1/discussions/")
#print(response,'\n')

df = pd.DataFrame(columns=["id", "titre", "contenu", "date"])

# Vérifie si la requête a réussi
if response.status_code == 200:
    # Récupère les données de la réponse
    data = response.json()
    print("Requête effectuée avec succès !\n")
    
    for discussion in data['data']:
        discussion = discussion['discussion']
        #print(discussion,'\n')
        #print(discussion[0])
        #[print(x) for x in discussion]
        for elem in discussion:
            mon_dict = elem
            #print(mon_dict) 
            print('ID:\n', mon_dict["posted_by"]["id"])
            print('Contenu:\n', mon_dict["content"])
            print('Date:\n', mon_dict["posted_on"],'\n')
            print('\n/////////////////////////////////////////////////////////////////////////////////////////\n')
        
        #current_df = pd.DataFrame(columns=["id", "titre", "contenu", "date"], data=[discussion["id"], discussion["title"], discussion["content"], discussion["posted_on"]])
        #df = pd.concat([df, current_df]) 
            
else:
    # Affiche un message d'erreur si la requête a échoué
    print("La requête a échoué avec le code d'erreur : ", response.status_code)
```
